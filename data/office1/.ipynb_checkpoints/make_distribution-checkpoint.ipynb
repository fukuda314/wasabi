```python
import os
import glob
from my_module.output_mod import output
import numpy as np
import pandas as pd
```

```python
current_dir_path = os.getcwd()
output_path = os.path.join(current_dir_path,'data')
```

```python
dir_list = glob.glob(output_path+'/*'+'/BoxCount.csv')#BoxCount.csvが入っているパス名探索
num_dir = len(dir_list)
```

```python
for path in range(num_dir):
    df_BoxCount = pd.read_csv(dir_list[path])
    """
    boxの分布を取得
    """
    x_max = max(df_BoxCount.loc[:, "x"])
    y_max = max(df_BoxCount.loc[:, "y"])
    x_min = min(df_BoxCount.loc[:, "x"])
    y_min = min(df_BoxCount.loc[:, "y"])
    box_size_x = df_BoxCount.iloc[1,0] - df_BoxCount.iloc[0,0]
    num_box_x = int((x_max - x_min) / box_size_x) + 1
    box_size_y = df_BoxCount.iloc[num_box_x + 1,1] - df_BoxCount.iloc[0,1]
    num_box_y = int((y_max - y_min) / box_size_y) + 1
    
    distribution = np.zeros((num_box_y, num_box_x),dtype=np.float32)#単精度実数で初期化
    for j in range(num_box_x):
        for i in range(num_box_y):
            distribution[i, j] = 1 - (np.exp(-df_BoxCount.iloc[(i * num_box_x) + j, 4] * 1e7 / (900/60)))#感染率計算
            
    os.chdir(os.path.dirname(dir_list[path]))
    output_result = output(output_name="risk_dist.csv", data=distribution, header_switch=False)#outputクラスのインスタンス作成
    output_result.output_csv()
```
